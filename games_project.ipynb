По некоторым названиям игр, для которых пропущен год выхода, год заметен в названии. Например, игры "Madden NFL 2004", "FIFA Soccer 2004" и другие. Также можно было бы поискать в таблице информацию о годе выпуска, посмотрев релизы одной и той же игры на разных платформах: возможно где-то год будет указан.
```python
# Удалим пропущенные значения в year_of_release и выполним проверку
data = data.dropna(subset=['year_of_release'])
data['year_of_release'].isna().sum()
```

```python
# Преобразуем данные в столбце year_of_release (год - целое число, а не дробное)
data['year_of_release'] = data['year_of_release'].astype('int')
# Проверим изменения
data.dtypes
```

```python
# Найдем уникальные значения в столбце user_score
data['user_score'].sort_values().unique()
```

***На основе полученных данных можно отметить, что:***

В уникальных значениях столбца с оценками пользователей мы видим аббревиатуру "tbd", что означает "to be determined" и переводится как "будет определено". То есть сейчас это значение является пропуском. Поэтому его можно заменить на NaN.

Что касается пропусков в данных, возможно, они появились, потому что некоторые игры не получили достаточное количество оценок от пользователей или вовсе не были ими оценены. Заменить пропуски на числа не получится, так как это будут совершенно неправильные оценки! Если бы мы поставили, например, медианные значения в пропущенных ячейках, у нас появились бы совсем недостоверные данные о большом количестве игр.

```python
# Заменим "tbd" в user_score на NaN
data['user_score'] = data['user_score'].replace('tbd', np.nan)
data['user_score'].sort_values().unique() 
```

```python
# Преобразуем данные в столбце user_score (оценки пользователей - дробное (вещественное) число, а не объект)
data['user_score'] = data['user_score'].astype('float')
# Проверим изменения
data.dtypes
```

```python
# Найдем уникальные значения в столбце critic_score
data['critic_score'].sort_values().unique()
```

***На основе полученных данных можно отметить, что:***

В столбце critic_score, как и в столбце user_score, заменить пропуски на числа не получится, потому что эти данные также будут недостоверны, и мы получим неправильные оценки. Возможно, пропуски в этом столбце появились, потому что некоторые игры не получили достаточное количество оценок от критиков или вовсе не были ими оценены.

### Проверьте гипотезы

#### Проверка гипотезы: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H0 - Средние пользовательские рейтинги платформ Xbox One и PC равны.

H1 - Средний пользовательский рейтинг платформы Xbox One отличается от среднего пользовательского рейтинга платформы PC.

```python
# Создадим переменную xone_user_score с оценками пользователей для платформы XOne для периода с 2014 по 2016 год
xone_user_score = actual_period_top_7[(actual_period_top_7['platform']=='XOne')&(actual_period_top_7['user_score'] >= 0)]['user_score']
xone_user_score
```

```python
# Посмотрим значения по xone_user_score
xone_user_score.describe()
```

```python
# Создадим переменную pc_user_score с оценками пользователей для платформы PC для периода с 2014 по 2016 год
pc_user_score = actual_period_top_7[(actual_period_top_7['platform']=='PC')&(actual_period_top_7['user_score'] >= 0)]['user_score']
pc_user_score
```

```python
# Посмотрим значения по pc_user_score
pc_user_score.describe()
```

```python
alpha = .01
results = st.ttest_ind(xone_user_score, pc_user_score)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

***На основе полученных данных можно отметить, что:***

Применив метод describe() к переменным xone_user_score и pc_user_score, мы заметили, что средние пользовательские рейтинги платформ Xbox One и PC  очень схожи - 6.594545 и  6.298361, медианные значения и вовсе равны - 6.800000.

Значение p-value равно 10%, что показывает, что это явно слишком большая вероятность, чтобы делать вывод о значимом различии между пользовательскими рейтингами платформ Xbox One и PC. Таким образом, можно сказать, что нулевая гипотеза подтверждается.

#### Проверка гипотезы: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные


Н0 - Средние пользовательские рейтинги жанров Action и Sports равны.

Н1 - Средний пользовательский рейтинг жанра Action отличается от среднего пользовательского рейтина жанра Sports.

```python
# Создадим переменную action_user_score с оценками пользователей для жанра Action с 2014 по 2016 год
action_user_score = actual_period_top_7[(actual_period_top_7['genre']=='Action')&(actual_period_top_7['user_score'] >= 0)]['user_score']
action_user_score
```

```python
# Посмотрим значения по action_user_score
action_user_score.describe()
```

```python
# Создадим переменную action_user_score с оценками пользователей для жанра Sports с 2014 по 2016 год
sports_user_score = actual_period_top_7[(actual_period_top_7['genre']=='Sports')&(actual_period_top_7['user_score'] >= 0)]['user_score']
sports_user_score
```

```python
# Посмотрим значения по sports_user_score
sports_user_score.describe()
```

```python
alpha = .01
results = st.ttest_ind(action_user_score, sports_user_score)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

***На основе полученных данных можно отметить, что:***

Применив метод describe() к переменным action_user_score и sports_user_score, мы заметили, что средние пользовательские рейтинги жанров Action и Sports составляют 6.783516 и 5.402804 соответственно и не являются похожими, медианные значения составляют 7.100000 и 5.700000 и так же не равны.

Значение p-value показывает, что средние пользовательские рейтинги жанров Action и Sports не являются равными. Таким образом, можно сказать, что нулевая гипотеза отвергается.
